```python
# гипотеза - количество предприятий в городах увеличилось с годами
# гипотеза - количество производственных предприятий сократилось
# гипотеза - наиболее устойчивая область экономики - добыча полезных ресурсов
```

```python
#количество предприятий в городах увеличилось с годами
settlements = df["settlement"].unique()
settlements
```

```python
def take_first(settlement_name):
    filtered = df.loc[df.loc[:, "settlement"] == settlement_name]
    #first_row = filtered.loc[filtered.loc[:,"year"] == filtered.loc[:,"year"].min()].reset_index(drop=True)
    first_row = filtered.loc[filtered.loc[:,"year"] == filtered.loc[:,"year"].min()].reset_index(drop=True)
    #print (first_row)
    return first_row

def take_last(settlement_name):
    filtered = df.loc[df.loc[:, "settlement"] == settlement_name]
    last_row = filtered.loc[filtered.loc[:,"year"] == filtered.loc[:,"year"].max()].reset_index(drop=True)
    #print(last_row)
    return last_row
```

```python
features = ["comp_electr", "comp_manufact", "comp_mining"]

for feature in features:
    grown_settlements = 0
    total_settlements = len(settlements)
    print(feature)
    for settlement in settlements:
            if take_first(settlement).loc[0, feature] < take_last(settlement).loc[0, feature]:
                grown_settlements += 1
    grown_part = grown_settlements/total_settlements
    print(f"Число {grown_part}")

print(f"Поселений с возросшим числом компаний")
```
